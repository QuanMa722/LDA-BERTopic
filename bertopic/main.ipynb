```python
with open(file="../data/lda4.txt", mode="r", encoding="utf-8") as f:
    comment_list_dict: list = f.readlines()
comment_list: list = []
for item in comment_list_dict:
    comment_list.append(eval(item)["评论内容"])
```

```python
comment_list[:10]
```

```python
with open("../data/stopwords.txt", mode="r", encoding="utf-8") as stop_file:
    stopwords = stop_file.readlines()
stopwords = [word.strip() for word in stopwords]
your_stopwords = []
```

```python
import jieba
corpus_list: list = []
for sentence in comment_list:
    corpus = jieba.lcut(sentence)
    corpus = [word for word in corpus if word not in stopwords]
    corpus = [word for word in corpus if word not in your_stopwords]
    corpus = [item for item in corpus if item != ' ']
    corpus = " ".join(corpus)
    print(corpus)
    corpus_list.append(corpus)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sentence_transformers import SentenceTransformer
from bertopic import BERTopic
from hdbscan import HDBSCAN
from umap import UMAP
import numpy as np
```

```python
embedding_model = SentenceTransformer(
    # 'paraphrase-multilingual-mpnet-base-v2',
      "thenlper/gte-base-zh" )
    # "thenlper/gte-large-zh"
embeddings = np.load('embedding.npy')
print(embeddings.shape)

umap_model = UMAP(
  n_neighbors=15,
  n_components=5,
  min_dist=0.0,
  metric='cosine',
  random_state=30 
)

hdbscan_model = HDBSCAN(
  min_cluster_size=35,
  min_samples=10,
  metric='euclidean'
)

vectorizer_model = CountVectorizer(stop_words=[])

topic_model = BERTopic(
  embedding_model=embedding_model,
  vectorizer_model=vectorizer_model,
  umap_model=umap_model,
  hdbscan_model=hdbscan_model,
)

```

```python
topics, probs = topic_model.fit_transform(corpus_list, embeddings=embeddings) 
topic_model.get_topic_info()
```

```python
topic_model.visualize_barchart()
```

```python
topic_model.visualize_topics()
```

```python
reduced_embeddings = UMAP(n_neighbors=10, n_components=2, min_dist=0.0, metric='cosine').fit_transform(embeddings)
topic_model.visualize_documents(corpus_list, reduced_embeddings=reduced_embeddings, hide_document_hover=True)
```

```python
hierarchical_topics = topic_model.hierarchical_topics(corpus_list)
topic_model.visualize_hierarchy(hierarchical_topics=hierarchical_topics)
```
